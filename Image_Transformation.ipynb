Linear Transformation

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('VIT-Bhopal-Logo.png', 0)
imginv = 255 - image

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
plt.title('Original', fontsize = 18)
plt.imshow(image, cmap='gray')
ax = fig.add_subplot(122)
plt.imshow(imginv, cmap='gray')
ax.set_title('Inverse Transform', fontsize=18)
```

Log Transformation

```python
image = cv2.imread('OpenCV_Logo.png', 0)
log = 0.6 * (np.log(1+ np.float32(image)))

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
plt.title('Original', fontsize = 18)
plt.imshow(image, cmap='gray')
ax = fig.add_subplot(122)
plt.imshow(log, cmap='gray')
ax.set_title('Log Transform', fontsize=18)
```

Gamma Transformation - Power Law Transformation

```python
image = cv2.imread('OpenCV_Logo.png', 0)
img_1 = image/255.0
img_power_law_transformation = cv2.pow(img_1, 1.8)
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(121)
plt.title('Original', fontsize = 18)
plt.imshow(image, cmap='gray')
ax = fig.add_subplot(122)
plt.imshow(img_power_law_transformation, cmap='gray')
ax.set_title('Power Law Transform', fontsize=18)
```
