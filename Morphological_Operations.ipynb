Import Statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Reading the image and setting up kernel

```python
img = cv2.imread('VIT-Bhopal-Logo.png')
kernel = np.ones((5,5),np.uint8)
plt.imshow(img)
```

Erosion

```python
erosion = cv2.erode(img, kernel)
plt.imshow(erosion)
plt.show()
```

Dilation

```python
dilation = cv2.dilate(img, kernel)
plt.imshow(dilation)
```
