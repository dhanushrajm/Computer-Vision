Import Statements

```python
import cv2
from matplotlib import pyplot as plt #used to display the images
```

Reading the image

```python
img = cv2.imread('/content/VIT-Bhopal-Logo.png')
name = 'VIT Bhopal Logo Image'
plt.title(name)
plt.imshow(img)
```

Flipping the image horizontally

```python
horizontal_flip = cv2.flip(img, 1)
name = 'VIT Bhopal Logo image Horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

Flipping the image vertially

```python
vertical_flip = cv2.flip(img, 0)
name = 'VIT Bhopal Logo image Vertically'
plt.title(name)
plt.imshow(vertical_flip)
```

Flipping the image Vertically and Horizontally

```python
vertical_horizontal_flip = cv2.flip(img, -1)
name = 'VIT Bhopal Logo Image flipped Vertically and Horizontally'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```
