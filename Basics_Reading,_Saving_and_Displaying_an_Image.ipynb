Importing libraries

```python
import numpy as np
from matplotlib import pyplot as plt
```

Reading an image

```python
import cv2

image = cv2.imread("/content/VIT-Bhopal-Logo.png")
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB )
plt.title('BGR to RGB - Original Image')
plt.imshow(image_rgb)
```

Saving the GrayScale Image

```python
gray_image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY )

#Saving using imwrite()
cv2.imwrite('Output2.jpeg', gray_image)

plt.title("Gray-Scale")
plt.imshow(gray_image)
```

Loading and Showing the image

```python
#Loading the image
img = cv2.imread("Output2.jpeg")

#Displaying the image
plt.title("Saved Image")
plt.imshow(img)
```
