```python
import random
import cv2
from matplotlib import pyplot as plt
```

Function Definition

Function add_noise() for adding random Black and While Pixels

This function will create a sprinked salt and pepper effect on the picture

```python
def add_noise(img):
    row, col = img.shape
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        ycord = random.randint(0, row - 1)
        xcord = random.randint(0, col - 1)
        img[ycord][xcord] = 255
    number_of_pixels = random.randint(300, 10000)
    for i in range(number_of_pixels):
        ycord = random.randint(0, row - 1)
        xcord = random.randint(0, col - 1)
        img[ycord][xcord] = 0
        
    return img
```

Reading the image

```python
image = cv2.imread('OpenCV_Logo.png')
plt.imshow(image)
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

Creating salt and pepper effect on the image

```python
img = cv2.imread('OpenCV_Logo.png', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('Logo_with_noise.jpg', add_noise(img))
noise_image = cv2.imread('Logo_with_noise.jpg')
plt.imshow(noise_image)
plt.show()
```

Enhancing the Image

```python
median = cv2.medianBlur(noise_image, 5)
plt.imshow(median)
plt.show()
```
